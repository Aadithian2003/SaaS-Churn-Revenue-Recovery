```python
import pandas as pd
import numpy as np
import re
from datetime import datetime, timedelta
```

```python
# 1. Generate Synthetic Messy Data
np.random.seed(42)
n_rows = 5000

data = {
    'transaction_id': range(1, n_rows + 1),
    'customer_id': np.random.randint(1000, 1500, n_rows),
    'amount': np.random.choice(['$19.99', '19.99 USD', '€18.50', '€18.50', '$49.99', '49.99 USD'], n_rows),
    'status': np.random.choice(['Success', 'Success', 'Success', 'Failed', 'Refunded'], n_rows),
    'event_date': [datetime(2023, 1, 1) + timedelta(days=np.random.randint(0, 365)) for _ in range(n_rows)],
    'last_login': [datetime(2023, 1, 1) + timedelta(days=np.random.randint(0, 365)) for _ in range(n_rows)]
}

df = pd.DataFrame(data)

# Injecting "The Mess": Duplicating 100 rows to simulate system glitches
df = pd.concat([df, df.sample(100)], ignore_index=True)

# Injecting Missing Values (NaNs)
# 5% of amounts are missing (system error)
df.loc[df.sample(frac=0.05).index, 'amount'] = np.nan

# 10% of last_login dates are missing (users who never logged in)
df.loc[df.sample(frac=0.10).index, 'last_login'] = np.nan
```

```python
# Identify duplicates based on ID, date, and amount
initial_count = len(df)
df = df.drop_duplicates(subset=['customer_id', 'event_date', 'amount', 'status'], keep='first')

print(f"Removed {initial_count - len(df)} duplicate system logs.")
```

```python
def normalize_revenue(row):
    val = str(row['amount'])
    if val == 'nan':
        return 0.0
    
    # Extract only the numbers and decimals
    numeric_part = re.sub(r'[^\d.]', '', val)
    if not numeric_part:
        return 0.0
    
    amount = float(numeric_part)
    
    # Apply exchange rate logic (Assuming 1 EUR = 1.08 USD)
    if '€' in val or 'EUR' in val:
        return round(amount * 1.08, 2)
    return amount

df['amount_usd'] = df.apply(normalize_revenue, axis=1)
```

```python
# Fill missing last_login with a placeholder or 'Never'
# For calculation purposes, we can set them to the event_date (signup date) 
# or a very old date to show they are inactive.
df['last_login'] = df['last_login'].fillna(pd.Timestamp('2000-01-01'))

# Flag 'Zombie Users'
df['is_zombie'] = df['last_login'] == pd.Timestamp('2000-01-01')
```

```python
# Create a flag for failed payments that haven't been recovered
df['is_failed_payment'] = df['status'] == 'Failed'

# Summary of the mess cleaned
print("--- Data Cleaning Summary ---")
print(f"Total Records: {len(df)}")
print(f"Total Revenue Lost to Failures: ${df[df['is_failed_payment']]['amount_usd'].sum():,.2f}")
```

```python
df.to_csv('cleaned_transactions.csv', index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the visual style
sns.set_theme(style="whitegrid")
```

```python
# 1. Define the cohort month (the first month a user had a 'Success')
df_success = df[df['status'] == 'Success'].copy()
df_success['order_month'] = df_success['event_date'].dt.to_period('M')
df_success['cohort_month'] = df_success.groupby('customer_id')['event_date'].transform('min').dt.to_period('M')

# 2. Calculate the 'age' of the customer (Cohort Index)
df_success['cohort_index'] = (df_success['order_month'].astype(int) - df_success['cohort_month'].astype(int))

# 3. Pivot the data
cohort_data = df_success.groupby(['cohort_month', 'cohort_index'])['customer_id'].nunique().reset_index()
cohort_pivot = cohort_data.pivot(index='cohort_month', columns='cohort_index', values='customer_id')

# 4. Convert to percentages (Retention Rate)
retention_matrix = cohort_pivot.divide(cohort_pivot.iloc[:,0], axis=0)

# 5. Plot the Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(retention_matrix, annot=True, fmt='.0%', cmap='Blues')
plt.title('Monthly Retention Cohorts')
plt.show()
```

```python
# Group by month and status
rev_summary = df.groupby([df['event_date'].dt.to_period('M'), 'status'])['amount_usd'].sum().unstack().fillna(0)

# Plotting the risk
rev_summary[['Success', 'Failed']].plot(kind='bar', stacked=True, figsize=(10,6), color=['#2ecc71', '#e74c3c'])
plt.title('Revenue: Successful vs. Failed (Involuntary Churn Risk)')
plt.ylabel('Revenue (USD)')
plt.xlabel('Month')
plt.legend(title='Payment Status')
plt.xticks(rotation=45)
plt.show()
```

```python


# 1. Define 'Churn' for this visual (e.g., no success in the last 60 days)
latest_date = df['event_date'].max()
last_success = df[df['status'] == 'Success'].groupby('customer_id')['event_date'].max()
df['is_churned'] = df['customer_id'].map(lambda x: (latest_date - last_success.get(x, latest_date)).days > 60)

# 2. Calculate inactivity
df['days_since_login'] = (latest_date - df['last_login']).dt.days

# 3. Plot the Correlation
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='days_since_login', hue='is_churned', fill=True, common_norm=False, palette='viridis')

plt.title('Churn Correlation: Days Since Last Login')
plt.xlabel('Days Inactive')
plt.ylabel('Density of Users')
plt.show()
```
